# Outliers analysis
## Using DBSCAN to detect outliers 
___

```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
from sklearn.cluster import DBSCAN
from matplotlib import rcParams
from collections import Counter

%matplotlib inline
rcParams['figure.figsize'] = 7.5 , 6
sns.set_style('whitegrid')

df = sns.load_dataset('iris')
cols = ['comp da cepa','larg  da cepa','comp da pétala','larg da pétala', 'espécie']

df.columns = cols
X = df[cols[:3]].values
y = df['espécie'].values
```

### Training the model to identify outliers

```python
model = DBSCAN(eps = 0.8, min_samples=19).fit(X)
print(model)
```

#### Viewing results

```python
outliers_df = pd.DataFrame(X)
print(Counter(model.labels_))
print(outliers_df[model.labels_==-1])
```

```python
color = model.labels_
plt.scatter(X[:,2], X[:,1], c=color, s=120)
plt.xlabel('lengh petal')
plt.ylabel('Width cepa')
plt.title ('DBSCAN to detect outlier')
plt.show()
```
