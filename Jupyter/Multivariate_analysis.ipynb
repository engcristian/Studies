# Analyzing Outliers
## Multivariate analysis to detect outliers
___

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
from matplotlib import rcParams

%matplotlib inline
rcParams['figure.figsize'] = 5, 4
sns.set_style('whitegrid')

df = sns.load_dataset('iris')
df.columns = ['comp da cepa', 'larg da cepa', 'comp da pétala','larg da pétala', 'espécie']
```

### Analysing boxplot visually

```python
sns.boxplot(x= 'espécie', y='comp da cepa', data =df, palette = 'hls')
```

### Looking for the scatterplot matrix

```python
sns.pairplot(df, hue= 'espécie', palette='hls')
```
